```python
x = np.random.chisquare(12, 1000)
plt.hist(x, bins=100,alpha=0.5,edgecolor='black')
plt.ylabel("N per bin")
plt.xlabel("x")
plt.savefig("cj2-10.pdf")
plt.show()
```
