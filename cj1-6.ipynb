```python
%matplotlib inline
import scipy as sp
import numpy as np
import matplotlib.pyplot as plt
from astropy.table import Table
```

define a table of 1000 values for x, ranging from 0 to 2*Pi

```python
npoints = 1000
x = np.linspace(0,2*np.pi,npoints)
```

calculate the sin(x) for each value in the x table and store in a table of values for y 

```python
def sin_x():
    y = np.sin(x)
    return y
```

calculate the cos(x) for each value in the x table and store in a table of values for z 

```python
def cos_x():
    z = np.cos(x)
    return z
```

define an astropy Table using the x, y, and z tables as columns, and create a loop to print the row values based on iteration count

(I opted to rewrite it to iterate once and print 10 rows, rather than 
1 row over 10 iterations. This prints the column names once)

```python
data = Table(data=[x,sin_x(),cos_x()], names=["x", "sin(x)", "cos(x)"])

for i, row in enumerate(data):
    #if i >= 10:
    #    break
    #print(data[i])  
    if i == 0:
        print(data[:10])
    else:
        break
```
